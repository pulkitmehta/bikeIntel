```python
for station in Stations:
    path = "../bikeIntel data/Stationwise bookings/"+station
    station_data = pd.read_csv(path+"/"+station+".csv")
    hrs = station_data['hh'].values
    hrs_1 = []
    
    for hr in hrs:

        if hr>=0 and hr<3:
            hr = '0-3'

        elif hr>=3 and hr<6:
            hr = '3-6'

        elif hr>=6 and hr<9:
            hr = '6-9'

        elif hr>=9 and hr<12:
            hr = '9-12'

        elif hr>=12 and hr<15:
            hr = '12-15'

        elif hr>=15 and hr<18:
            hr = '15-18'

        elif hr>=18 and hr<21:
            hr = '18-21'

        elif hr>=21 and hr<24:
            hr = '21-24'
    
        hrs_1.append(hr)
        
    hrs = pd.Series(hrs_1, name="hh")
    
    pd.concat([station_data.iloc[:,:3],hrs, station_data['Day'].astype(str)], axis = 1).to_csv(path+"/"+station+"_hr_binned.csv", index = False)
    
    print(station+"_hr_binned.csv processed successfully...")
```
